```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
datasets = pd.read_csv("test.csv")
```

```python
datasets.head(5)
```

```python
(datasets.isnull().sum()/187531)*100
```

```python
datasets.shape
```

```python
(datasets.isnull().sum().sum()/(datasets.shape[0]*datasets.shape[1]))*100
```

```python
feature = datasets[["GrLivArea","FullBath","BedroomAbvGr"]]
target = datasets["YrSold"]
```

```python
features = feature.fillna(0)
```

```python
X_train , X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print("MEAN SQUARED ERROR: ",mse)
print("r2 score: ",r2)
```

```python
coefficients = pd.DataFrame({'FEATURE':features.columns,'coefficent':model.coef_})
```

```python
print(coefficients)
```
